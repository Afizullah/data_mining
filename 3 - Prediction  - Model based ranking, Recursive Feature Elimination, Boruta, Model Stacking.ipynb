#### The feature mean distance of the 30 nearest POI

```python
df_tourisme = pd.read_csv("paris_poi.csv", index_col=0)
```

```python
df_tourisme
```

```python
def nearest(df_, i, df_t, n:int):
    
    """
    return the index in df_t of the "n" neareast and their distance
    in the form (distances, indice)
    """
    x = np.array([df_['lat'][i], df_['lon'][i]])
    ys = np.array(list(zip(df_t['lat'],  df_t['lon'])))
    #distances in km
    distances = np.linalg.norm(x - ys, axis=1) * 77.14017896147689
    indices = distances.argsort()[:n]
    
    return distances[indices], indices
    
```

```python
df_v3["30 nearest"]=[nearest(df_v3,i,df_tourisme,30) for i in range(len(df_v3))]
```

```python
df_v3["30 nearest mean (km)"]=[i[0].mean()for i in df_v3["30 nearest"]]
```

```python
from sklearn.model_selection import cross_val_predict, GridSearchCV
```

```python
def grid_search(X,Y,estimator, param,cv):
    gsc=GridSearchCV(estimator=estimator,
                     param_grid=param,
                     cv=cv,
                     scoring='neg_mean_squared_error',
                     n_jobs=-1)
    result=gsc.fit(X,Y)
    best_params=result.best_params_
    
    return result, best_params
```

#### Stacking

```python
# Fonction to preprare the dataset for the second regression

from sklearn.model_selection import KFold

def get_stacking(clf, X_train, Y_train, X_test, n_folds):
    
    train_num, test_num = X_train.shape[0], X_test.shape[0]
    
    second_level_train_set = np.zeros((train_num,))
    second_level_test_set = np.zeros((test_num,))
    test_nfolds_sets = np.zeros((test_num, n_folds))
    kf = KFold(n_splits=n_folds)
    
    for i,(train_index, test_index) in enumerate(kf.split(X_train)):
        x_tra, y_tra = X_train[train_index], Y_train[train_index]
        x_tst, y_tst =  X_train[test_index], Y_train[test_index]

        clf.fit(x_tra, y_tra)

        second_level_train_set[test_index] = clf.predict(x_tst)
        test_nfolds_sets[:,i] = clf.predict(X_test)
        i+=1
        

    second_level_test_set[:] = test_nfolds_sets.mean(axis=1)
    
    return second_level_train_set, second_level_test_set
```

```python
# 3 selected basic estimators

clfs = [RandomForestRegressor(n_estimators = 300, max_depth=20,n_jobs=-1,random_state=False, verbose=False),
        GradientBoostingRegressor(n_estimators = 400, learning_rate=0.1,loss='ls',random_state=False),
        ExtraTreesRegressor(n_estimators = 300,max_depth=19,n_jobs=-1,random_state=False)]
```

```python
import seaborn as sns
```

```python
# Cross validation Stacking model

from sklearn.linear_model import LinearRegression

k_f=KFold(n_splits=3)
y_pred_i = np.zeros(X.shape[0])

for train_index, test_index in k_f.split(X):
    
    print(test_index)
    X_train_i, Y_train_i = np.array(X)[train_index], np.array(Y)[train_index]
    X_test_i, Y_test_i =  np.array(X)[test_index], np.array(Y)[test_index]

    train_sets_i = []
    test_sets_i = []
    
    for clf in clfs:
        train_set_i, test_set_i = get_stacking(clf=clf, X_train=X_train_i, Y_train=Y_train_i, X_test=X_test_i,n_folds=5)
        train_sets_i.append(train_set_i)
        test_sets_i.append(test_set_i)
        
    meta_train_i = np.concatenate([result_set.reshape(-1,1) for result_set in train_sets_i], axis=1)
    meta_test_i = np.concatenate([y_test_set.reshape(-1,1) for y_test_set in test_sets_i], axis=1)
    
    df_=pd.DataFrame(meta_train_i)
    
    matrix = np.triu(df_.corr())
    ax=sns.heatmap(df_.corr(), annot=True, mask=matrix,annot_kws={'size':20})
    plt.show()

    model=LinearRegression()
    model.fit(meta_train_i,Y_train_i)
    y_pred_i[test_index]=model.predict(meta_test_i)
        
```
